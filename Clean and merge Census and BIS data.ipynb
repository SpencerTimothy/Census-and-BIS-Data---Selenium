```python
# Import necessary libraries
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
# Set up the Selenium web driver (replace 'path_to_web_driver' with the actual path to your web driver)
driver = webdriver.Chrome()

# Load the webpage with the table
driver.get('https://www.bls.gov/respondents/mwr/electronic-data-interchange/appendix-d-usps-state-abbreviations-and-fips-codes.htm')

# Find the table element
table = driver.find_element(By.XPATH,'//table')

# Extract the rows from the table
rows = table.find_elements(By.XPATH,'.//tr')

# Iterate over the rows and extract the desired data for data1
data1 = []
for row in rows:
    cells = row.find_elements(By.XPATH,'.//td')
    state = cells[0].text
    postal_abbr = cells[1].text
    fips_code = cells[2].text
    data1.append((state, postal_abbr, fips_code))

# Iterate over the rows and extract the desired data for data2
data2 = []
for row in rows:
    cells = row.find_elements(By.XPATH,'.//td')
    state = cells[3].text
    postal_abbr = cells[4].text
    fips_code = cells[5].text
    data2.append((state, postal_abbr, fips_code))

# 2000 data retrieval:

# Load the webpage and extract data for the 110th land area
driver.get('https://www2.census.gov/geo/relfiles/cd110th/natl_code/land_cd110_natl.txt')
land_area_110 = driver.find_element(By.TAG_NAME, 'pre').text
lines = land_area_110.split('\n')
header = lines[1].split(',')
to_df = [line.split(',') for line in lines[2:]]
land_area_110_df = pd.DataFrame(to_df, columns=header)

# Load the webpage and extract data for the 109th land area
driver.get('https://www2.census.gov/geo/relfiles/cd109th/natl_code/ua_cd109_natl.txt')
land_area_109 = driver.find_element(By.TAG_NAME, 'pre').text
lines = land_area_109.split('\n')
header = lines[1].split(',')
to_df = [line.split(',') for line in lines[2:]]
land_area_109_df = pd.DataFrame(to_df, columns=header)

# Load the webpage and extract data for the 110th zip code tabulation area
driver.get('https://www2.census.gov/geo/relfiles/cd110th/natl_code/zcta_cd110_natl.txt')
zip_code_110 = driver.find_element(By.TAG_NAME, 'pre').text
lines = zip_code_110.split('\n')
header = lines[1].split(',')
to_df = [line.split(',') for line in lines[2:]]
zip_code_110_df = pd.DataFrame(to_df, columns=header)

# Load the webpage and extract data for the 109th zip code tabulation area
driver.get('https://www2.census.gov/geo/relfiles/cd109th/natl_code/zcta_cd109_natl.txt')
zip_code_109 = driver.find_element(By.TAG_NAME, 'pre').text
lines = zip_code_109.split('\n')
header = lines[1].split(',')
to_df = [line.split(',') for line in lines[2:]]
zip_code_109_df = pd.DataFrame(to_df, columns=header)

# 2010 data retrieval:

# Load the webpage and extract data for the 116th land area
driver.get('https://www2.census.gov/geo/relfiles/cdsld18/natl/natl_landarea_cd_delim.txt')
land_area_116 = driver.find_element(By.TAG_NAME, 'pre').text
lines = land_area_116.split('\n')
header = lines[1].split(',')
to_df = [line.split(',') for line in lines[2:]]
land_area_116_df = pd.DataFrame(to_df, columns=header)

# Load the webpage and extract data for the 115th land area
driver.get('https://www2.census.gov/geo/relfiles/cdsld16/natl/natl_landarea_cd_delim.txt')
land_area_115 = driver.find_element(By.TAG_NAME, 'pre').text
lines = land_area_115.split('\n')
header = lines[1].split(',')
to_df = [line.split(',') for line in lines[2:]]
land_area_115_df = pd.DataFrame(to_df, columns=header)

# Load the webpage and extract data for the 113th land area
driver.get('https://www2.census.gov/geo/relfiles/cdsld13/natl/natl_landarea_cd_delim.txt')
land_area_113 = driver.find_element(By.TAG_NAME, 'pre').text
lines = land_area_113.split('\n')
header = lines[1].split(',')
to_df = [line.split(',') for line in lines[2:]]
land_area_113_df = pd.DataFrame(to_df, columns=header)

# Load the webpage and extract data for the 116th zip code tabulation area
driver.get('https://www2.census.gov/geo/relfiles/cdsld18/natl/natl_zccd_delim.txt')
zip_code_116 = driver.find_element(By.TAG_NAME, 'pre').text
lines = zip_code_116.split('\n')
header = lines[1].split(',')
to_df = [line.split(',') for line in lines[2:]]
zip_code_116_df = pd.DataFrame(to_df, columns=header)

# Load the webpage and extract data for the 115th zip code tabulation area
driver.get('https://www2.census.gov/geo/relfiles/cdsld16/natl/natl_zccd_delim.txt')
zip_code_115 = driver.find_element(By.TAG_NAME, 'pre').text
lines = zip_code_115.split('\n')
header = lines[1].split(',')
to_df = [line.split(',') for line in lines[2:]]
zip_code_115_df = pd.DataFrame(to_df, columns=header)

# Load the webpage and extract data for the 113th zip code tabulation area
driver.get('https://www2.census.gov/geo/relfiles/cdsld13/natl/natl_zccd_delim.txt')
zip_code_113 = driver.find_element(By.TAG_NAME, 'pre').text
lines = zip_code_113.split('\n')
header = lines[1].split(',')
to_df = [line.split(',') for line in lines[2:]]
zip_code_113_df = pd.DataFrame(to_df, columns=header)

# Close the web driver
driver.quit()
```

```python
# Create a DataFrame 'states1' with data from 'data1' and specify column names
states1 = pd.DataFrame(data1, columns=['state', 'postal_abbr', 'fips_code'])

# Remove the first two rows from 'states1'
states1 = states1.iloc[2:]

# Reset the index of 'states1' to reindex from 0 to N-1
states1 = states1.reset_index(drop=True)

# Create a DataFrame 'states2' with data from 'data2' and specify column names
states2 = pd.DataFrame(data2, columns=['state', 'postal_abbr', 'fips_code'])

# Remove the last row from 'states2'
states2 = states2.drop(states2.index[-1])

# Reset the index of 'states2' to reindex from 0 to N-1
states2 = states2.reset_index(drop=True)
```

```python
# Concatenate 'states1' and 'states2' DataFrames along axis 0 (row-wise)
states = pd.concat([states1, states2], axis=0)

# Reset the index of the resulting DataFrame to have a continuous integer index starting from 0
states = states.reset_index(drop=True)
```

```python
# 2000
# Merge 'states' DataFrame with 'land_area_109_df' DataFrame based on the 'fips_code' in 'states' and 'STATE' in 'land_area_109_df'
land_area_109_raw = pd.merge(states, land_area_109_df, left_on='fips_code', right_on='STATE')

# Merge 'states' DataFrame with 'land_area_110_df' DataFrame based on the 'fips_code' in 'states' and 'STATE' in 'land_area_110_df'
land_area_110_raw = pd.merge(states, land_area_110_df, left_on='fips_code', right_on='STATE')

# Merge 'states' DataFrame with 'zip_code_109_df' DataFrame based on the 'fips_code' in 'states' and 'STATE' in 'zip_code_109_df'
zip_code_109_raw = pd.merge(states, zip_code_109_df, left_on='fips_code', right_on='STATE')

# Merge 'states' DataFrame with 'zip_code_110_df' DataFrame based on the 'fips_code' in 'states' and 'STATE' in 'zip_code_110_df'
zip_code_110_raw = pd.merge(states, zip_code_110_df, left_on='fips_code', right_on='STATE')


# 2010
# Merge 'states' DataFrame with 'land_area_116_df' DataFrame based on the 'fips_code' in 'states' and 'State' in 'land_area_116_df'
land_area_116_raw = pd.merge(states, land_area_116_df, left_on='fips_code', right_on='State')

# Merge 'states' DataFrame with 'land_area_115_df' DataFrame based on the 'fips_code' in 'states' and 'State' in 'land_area_115_df'
land_area_115_raw = pd.merge(states, land_area_115_df, left_on='fips_code', right_on='State')

# Merge 'states' DataFrame with 'land_area_113_df' DataFrame based on the 'fips_code' in 'states' and 'State' in 'land_area_113_df'
land_area_113_raw = pd.merge(states, land_area_113_df, left_on='fips_code', right_on='State')

# Merge 'states' DataFrame with 'zip_code_116_df' DataFrame based on the 'fips_code' in 'states' and 'State' in 'zip_code_116_df'
zip_code_116_raw = pd.merge(states, zip_code_116_df, left_on='fips_code', right_on='State')

# Merge 'states' DataFrame with 'zip_code_115_df' DataFrame based on the 'fips_code' in 'states' and 'State' in 'zip_code_115_df'
zip_code_115_raw = pd.merge(states, zip_code_115_df, left_on='fips_code', right_on='State')

# Merge 'states' DataFrame with 'zip_code_113_df' DataFrame based on the 'fips_code' in 'states' and 'State' in 'zip_code_113_df'
zip_code_113_raw = pd.merge(states, zip_code_113_df, left_on='fips_code', right_on='State')
```
